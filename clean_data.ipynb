```python
# drop off NaN value in sunrise_sunset, civil_twilight nautical_twilight and astronomical_twilight
clean1_df = df[df['Sunrise_Sunset'].notnull() | df['Civil_Twilight'].notnull() | df['Nautical_Twilight'].notnull() | df['Astronomical_Twilight'].notnull()]
```

```python
# drop off NaN in weather_condtion
clean2_df = clean1_df[clean1_df['Weather_Condition'].notnull()]
clean2_df
```

```python
#drop off number, don't know what this is, TMC seems like the code to describe the accident, end lat, end lag
#drop off description,for city, there are too many cities, or we can divide city into a few categories, but i decide to drop it
clean3_df = clean2_df.drop(columns = ['ID','TMC','Description','Number','Street','Zipcode','Precipitation(in)','Wind_Chill(F)','End_Lat','End_Lng','End_Time','Country'])
clean3_df.dtypes
```

# 1.based on the stardard of dropping features with less onther 300000 recodes we have clean3_df, now i am wondering if we 
# need to drop wind speed also, but i think this is really relevant to our prediction, i want to leave it as is

```python
# start time grouping, when category time, i want to include monty which could reflect seasonality and hour, which reflects
# the intensity of traffic.

import datetime as dt
clean3_df['Start_Time'] = pd.to_datetime(clean3_df['Start_Time']).dt.strftime('%m-%H')
```

```python
#after check the density of the city, i use the group of 0-20, 20-100,100-500, 500-1000,1000-3000,3000-6000,6000-15000,and leave
# the city has 15000 more accidents
# with total of 7 groups + each city more than 15000
G0_20 = clean3_df['City'].value_counts()[clean3_df['City'].value_counts()<=20].index
G20_100 = clean3_df['City'].value_counts()[(clean3_df['City'].value_counts()>20) & (clean3_df['City'].value_counts()<=500)].index
G100_500 = clean3_df['City'].value_counts()[(clean3_df['City'].value_counts()>100) & (clean3_df['City'].value_counts()<=500)].index
G500_1000 = clean3_df['City'].value_counts()[(clean3_df['City'].value_counts()>500) & (clean3_df['City'].value_counts()<=1000)].index
G1000_3000 = clean3_df['City'].value_counts()[(clean3_df['City'].value_counts()>1000) & (clean3_df['City'].value_counts()<=3000)].index
G3000_6000 = clean3_df['City'].value_counts()[(clean3_df['City'].value_counts()>3000) & (clean3_df['City'].value_counts()<=6000)].index
G6000_15000 = clean3_df['City'].value_counts()[(clean3_df['City'].value_counts()>6000) & (clean3_df['City'].value_counts()<=15000)].index

clean3_df['City'] = clean3_df['City'].apply(lambda x:'0-20' if x in G0_20 else x)
clean3_df['City'] = clean3_df['City'].apply(lambda x:'20_100' if x in G20_100 else x)
clean3_df['City'] = clean3_df['City'].apply(lambda x:'100_500' if x in G100_500 else x)
clean3_df['City'] = clean3_df['City'].apply(lambda x:'500-1000' if x in G500_1000 else x)
clean3_df['City'] = clean3_df['City'].apply(lambda x:'1000_3000' if x in G1000_3000 else x)
clean3_df['City'] = clean3_df['City'].apply(lambda x:'3000-6000' if x in G3000_6000 else x)
clean3_df['City'] = clean3_df['City'].apply(lambda x:'6000-15000' if x in G6000_15000 else x)
```

```python
clean3_df['County'].value_counts()[30:300]
```

```python
# same method applied to county 
# there are two categories to use, accident under 1000, and 1000-8000
G1000 = clean3_df['County'].value_counts()[clean3_df['County'].value_counts()<=1000].index
G8000 = clean3_df['County'].value_counts()[(clean3_df['County'].value_counts()>1000) & (clean3_df['County'].value_counts()<=8000)].index
clean3_df['County'] = clean3_df['County'].apply(lambda x:'0-1000' if x in G1000 else x)
clean3_df['County'] = clean3_df['County'].apply(lambda x:'1000-8000' if x in G8000 else x)
```

```python
# At this port to keep wind_speed,clean5_df will be the final dataset to build our model
clean5_df = clean4_df.dropna()
clean5_df.notnull().sum()
```

```python
list1 = clean5_df['Weather_Condition'].value_counts()[clean5_df['Weather_Condition'].value_counts() < 561].index
clean5_df['Weather_Condition'] = clean5_df['Weather_Condition'].apply(lambda x: 'Other_Condition' if x in list1 else x )
```

```python
list2 = clean5_df['State'].value_counts()[clean5_df['State'].value_counts() < 3500].index
clean5_df['State'] = clean5_df['State'].apply(lambda x: 'Other_State' if x in list2 else x )
# somehow  the 27076 records have unidentified data ?
clean5_df = clean5_df[clean5_df.index != 27076]
```

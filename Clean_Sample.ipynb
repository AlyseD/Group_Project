```python
import pandas as pd
import datetime as dt
```

```python
#Bring in original sample from CSV
df=pd.read_csv("./US_Accidents_June20.csv")
df.head(5)
```

```python
df.isnull().sum()
```

```python
#Drop columns with signifiant amounts of null values
df_new=df.drop(['TMC', 'End_Lat', 'End_Lng', 'Number', 'Wind_Chill(F)', 'Precipitation(in)'], axis=1)
df_new
```

```python
df_new.isnull().sum()
```

```python
#Drop rows with null values
df_clean=df_new.dropna()
df_clean
```

```python
df_clean.isnull().sum()
```

```python
df_clean.dtypes
```

```python
#Select 500k sample rows
df_sample = df_clean.sample(n = 500000)
df_sample
```

```python
df_sample.to_csv(r'C:\BootcampUCB\Final Project\df_sample_full.csv', index = True)
```

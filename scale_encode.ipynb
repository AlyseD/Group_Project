```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf
import pandas as pd
import datetime as dt
```

```python
Data = pd.read_csv('cleaned_Data.csv')
```

```python
en_list = Data.dtypes[(Data.dtypes == "object") | (Data.dtypes == "bool")].index.tolist()

```

```python

encoder = OneHotEncoder(sparse = False)
encode_df = pd.DataFrame(encoder.fit_transform(Data[en_list][:1000000]))

encode_df.columns = encoder.get_feature_names(en_list)
encode_df.head()
```

```python
X = Data
y = Data['Severity']
Data.columns
```

```python
# not using the whole dataset,because my computer can't handle it
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 250000,train_size = 1000000,random_state = 1, stratify = y)
X_train
```

```python
encoder = OneHotEncoder(sparse = False)
encode_df = pd.DataFrame(encoder.fit_transform(X_train[en_list]),index = X_train.index)

encode_df.columns = encoder.get_feature_names(en_list)
encode_df.head()


encoder1 = OneHotEncoder(sparse = False)
encode_df_test = pd.DataFrame(encoder1.fit_transform(X_test[en_list]),index = X_test.index)

encode_df_test.columns = encoder1.get_feature_names(en_list)
encode_df
```

```python
X_train = X_train.merge(encode_df,left_index = True, right_index = True)
X_train.drop(columns = en_list,inplace = True)

X_test = X_test.merge(encode_df_test,left_index = True, right_index = True)
X_test.drop(columns = en_list,inplace = True)
X_train
```

```python
sc = StandardScaler()
sc.fit(X_train)
X_train_sc = sc.transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_df = pd.DataFrame(X_train_sc,columns =X_train.columns,index = X_train.index)
X_test_df = pd.DataFrame(X_test_sc,columns =X_test.columns, index = X_test.index)
```
